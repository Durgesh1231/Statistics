```python
# 1
"""
Statistics is the branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data.
"""

# 2
"""
There are two types of statistics: Descriptive and Inferential.

1. Descriptive Statistics: It summarizes or describes a set of data. Example: Using the mean, median, and mode to describe exam scores.
2. Inferential Statistics: It makes inferences or predictions about a population based on a sample. Example: Using sample data to predict the average income of a country's population.
"""

# 3
"""
There are two main types of data: Quantitative and Qualitative.

1. Quantitative Data: Numerical data that can be measured or counted. Example: Age, height, or weight.
2. Qualitative Data: Categorical data that represents characteristics or attributes. Example: Colors, gender, or nationality.
"""

# 5
"""
Levels of measurement refer to the different ways data can be categorized and measured. There are four levels:

1. Nominal: Data can only be categorized. Example: Gender (male/female).
2. Ordinal: Data can be ordered or ranked. Example: Education level (high school, undergraduate, graduate).
3. Interval: Data with meaningful intervals but no true zero. Example: Temperature in Celsius.
4. Ratio: Data with a true zero and meaningful intervals. Example: Weight or height.
"""

# 6
"""
Understanding the level of measurement is important because it determines the type of statistical analysis you can perform. For example, nominal data cannot be used for calculating means, while interval or ratio data can.
Example: You can't calculate the mean of "colors of cars" (nominal data), but you can calculate the mean height (ratio data).
"""

# 7
"""
Nominal data represents categories without any order, such as gender or color. Ordinal data, on the other hand, represents categories with a meaningful order, like educational level (e.g., high school, bachelor's, master's).
"""

# 8
"""
A box plot can be used to display data in terms of range. It shows the minimum, first quartile, median, third quartile, and maximum of the data.
"""

# 9
"""
Descriptive statistics summarize the data, while inferential statistics make predictions or generalizations about a population based on a sample.

Example of Descriptive Statistics: Finding the average test score of a class.
Example of Inferential Statistics: Using a sample of voters to predict the outcome of an election.
"""

# 10
"""
Common measures of central tendency are the mean, median, and mode. Common measures of variability include range, variance, and standard deviation.

1. Mean: The average of the data. Used to describe the central value of a dataset.
2. Median: The middle value in a dataset when arranged in order. Used when the data has outliers.
3. Mode: The most frequent value in a dataset. Useful for categorical data.
4. Range: The difference between the maximum and minimum values. Used to describe the spread of the data.
"""

# 4
"""
(i) Grading in exam: Qualitative (Ordinal)
(ii) Colour of mangoes: Qualitative (Nominal)
(iii) Height data of a class: Quantitative (Continuous)
(iv) Number of mangoes exported by a farm: Quantitative (Discrete)
"""
```
